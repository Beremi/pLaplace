```python
import scipy.io as sio
data = sio.loadmat("mesh_3D_level0.mat")
```

```python
elems = data["elems"]
coords = data["coords"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# create figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# plot vertices
ax.scatter(coords[:,0], coords[:,1], coords[:,2])

# plot edges
edges = []
for elem in elems:
    for i in range(4):
        for j in range(i+1, 4):
            edge = [elem[i]-1, elem[j]-1]
            edge.sort()
            if edge not in edges:
                edges.append(edge)
for edge in edges:
    ax.plot3D(*zip(*coords[edge]), color='k')

plt.show()
```
